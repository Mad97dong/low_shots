```python
names = []
images = []
image_size = (3, 128, 128)
fill_color = (255, 255, 255)
num_of_low_shots = 24
num_of_val = 8*8

for img in os.listdir('./pokemon/images/images'):
    im = Image.open('./pokemon/images/images/{}'.format(img))
    names.append(img.split('.')[0])
    im = im.convert("RGBA")
    if im.mode in ('RGBA', 'LA'):
        bg = Image.new(im.mode[:-1], image_size[1:], fill_color) # size = (128, 128, 3)
        bg.paste(im, im.split()[-1])
        im = bg

    images.append(np.asarray(im))
    if len(images) == 800: # crop at 800 images
        break
    
images = (np.asarray(images).astype(np.float32) / 255.0)
```

```python
batch_size = 2

class Pokeman(Dataset):
    def __init__(self, imgs, names):
        self.data = imgs
        self.label = names
        
        self.transform=transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
        ])
        
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return self.transform(self.data[idx]), self.label[idx], idx
    
train_set = Pokeman(images[:num_of_low_shots], names[:num_of_low_shots])
train_loader = torch.utils.data.DataLoader(
    train_set, batch_size=batch_size, 
    shuffle=True, num_workers=8, pin_memory=False,
)


val_set = Pokeman(images[-num_of_val:], names[-num_of_val:])
val_loader = torch.utils.data.DataLoader(val_set, shuffle=False, batch_size=1)
len(val_set)
```

```python
def project_l2_ball(z):
    """ project vectors in z onto the l2 unit norm ball"""
    return z / np.maximum(np.sqrt(np.sum(z**2, axis=1))[:, np.newaxis], 1)

def imsave(filename, array):
    im = Image.fromarray((array * 255).astype(np.uint8))
    im.save(filename)
```

```python
# DCGAN architecture
code_dim = 128
class Decoder(nn.Module):
    def __init__(self, code_dim, n_filter=64, out_channels=3):
        super(Decoder, self).__init__()
        self.code_dim = code_dim
        nf = n_filter
        self.dcnn = nn.Sequential(
            nn.ConvTranspose2d(code_dim, nf * 16, 4, 1, 0, bias=False),
            nn.BatchNorm2d(nf * 16), 
            nn.ReLU(True), # 512 * 4 * 4
            nn.ConvTranspose2d(nf * 16, nf * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nf * 8),   
            nn.ReLU(True), # 256 * 8 * 8
            nn.ConvTranspose2d(nf * 8, nf * 4, 4, 2, 1, bias=False), 
            nn.BatchNorm2d(nf * 4), 
            nn.ReLU(True), # 128 * 16 * 16
            nn.ConvTranspose2d(nf * 4, nf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nf * 2), 
            nn.ReLU(True), # 64 * 32 * 32
            nn.ConvTranspose2d(nf * 2, nf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nf), 
            nn.ReLU(True), # 32 * 64 * 64
            nn.ConvTranspose2d(nf, out_channels, kernel_size=4, stride=2, padding=1, bias=False),
            nn.Tanh(),
            # 3 * 128 * 128
        )

    def forward(self, code):
        return self.dcnn(code.view(code.size(0), self.code_dim, 1, 1))
```

```python
class LatentVariables(nn.Module):
    def __init__(self, Z):
        super(LatentVariables, self).__init__()
        self.Z = Parameter(Z)
        self.dim_Z = Z.size(0) # = number of low shots

    def forward(self, indices): # indices should be a list
        if max(indices) > self.dim_Z-1:
            print(indices)
        return self.Z[indices]
```

```python
Z = np.random.normal(size=(len(train_set), 1, code_dim)).astype(np.float32) # (number of low shots, 128)
Z = torch.tensor(project_l2_ball(Z), requires_grad=True).to(device)
Z.shape
```

```python
LV_Z = LatentVariables(Z)
pre_decoder = Decoder(code_dim)
```

```python
class CombinedModel(nn.Module):
    def __init__(self, LV_Z, Decoder, n_filters=64):
        super(CombinedModel, self).__init__()
        self.Z = LV_Z
        self.Decoder = Decoder

    def forward(self, index):
        code = self.Z(index)
        return self.Decoder(code)
```

```python
def weight_init(module):
    classname = module.__class__.__name__
    if classname.find('Conv')!=-1:
        module.weight.data.normal_(0.0, .02)
    elif classname.find('BatchNorm') != -1:
        module.weight.data.normal_(1.0, .01)
        module.bias.data.fill_(0.0)

g = CombinedModel(LV_Z, pre_decoder).to(device)
g.apply(weight_init)
print(g)
```

```python
import shutil

def save_ckp(state, is_best, checkpoint_path, best_model_path):
    f_path = checkpoint_path
    torch.save(state, f_path)
    if is_best:
        best_fpath = best_model_path
        shutil.copyfile(f_path, best_fpath)
        
def load_ckp(checkpoint_fpath, model_to_load, optimizer):
    checkpoint = torch.load(checkpoint_fpath)
    model_to_load.load_state_dict(checkpoint['state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer'])
    return model_to_load, optimizer, checkpoint['epoch']
```

```python
# prepare for pre-train
loss_fn = nn.MSELoss()

learning_rate = 1e-4
optimizer = torch.optim.Adam(g.parameters(), lr=learning_rate)

Xi_val, _, idx_val = next(iter(val_loader)) # Xi_val: [1, 3, 128, 128]
imsave('target.png', make_grid(Xi_val.cpu() / 2. + 0.5, nrow=8).numpy().transpose(1, 2, 0))
Xi_val.shape
```

```python
train_image_index = np.arange(0, len(train_loader))

def pre_train(start_epoch, n_epochs, train_loader, g, optimizer, pre_decoder, save_decoder):
    g.train()
    
    for epoch in range(start_epoch, n_epochs+1):
        losses = []
        progress = tqdm(total=len(train_loader), desc='epoch % 3d' % epoch)
        
        for i, (Xi, _, idx) in enumerate(train_loader):
    
            Xi = Variable(Xi.to(device))

            G_zi = g(idx.numpy())
            loss = loss_fn(G_zi, Xi)
        
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
#             y = Z[idx.numpy()].clone()
            Z[idx.numpy()] = torch.tensor(project_l2_ball(Z[idx.numpy()].data.cpu().numpy()), requires_grad=True).to(device)       
#             print(y==Z[idx.numpy()])
            
            losses.append(loss.item())
            progress.set_postfix({'loss': np.mean(losses[-99:])})
            progress.update()

        progress.close()
    
        # visualize reconstructions on training data
        G_zi = g(train_image_index)
    
        if epoch%30 == 0:
            imsave('./low_shots_pk/%s_epoch_%03d.png' % ('pk', epoch), 
                make_grid(G_zi.data.cpu() / 2. + 0.5, nrow=8).numpy().transpose(1, 2, 0))
        
    # save the network parameter 
    torch.save(pre_decoder.state_dict(), save_decoder)
    return g
```

```python
epoch_for_pre = 180
g = pre_train(1, epoch_for_pre, train_loader, g, optimizer, pre_decoder,
                       "./low_shots_pk/checkpoint/current_decoder_pre.pt")
```

```python
loss_fn = nn.MSELoss()

learning_rate = 5 * 1e-4
epoch_for_train_z = 1250 # 1250

for param in train_decoder.parameters():
    param.requires_grad = False

optimizer = torch.optim.Adam(train_g.parameters(), lr=learning_rate)

Xi_val, _, idx_val = next(iter(val_loader)) # Xi_val: [1, 3, 128, 128]
Xi_val = Variable(Xi_val).to(device)

plt.imshow(Xi_val.cpu().numpy().squeeze().transpose(1, 2, 0) / 2. + 0.5)
print(Xi_val.dtype, idx_val.numpy())
```

```python
def find_z(start_epoch, n_epochs, val_loader, val_idx, train_g, optimizer, target_z, train_decoder, train_z_path):

    train_g.train()
    for epoch in range(start_epoch, n_epochs+1):
        losses = []
        progress = tqdm(total=1, desc='epoch % 3d' % epoch) # for one target image
        
        G_z = train_g(val_idx)
        
        loss = loss_fn(G_z, Xi_val)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # project z into the unit ball
        Z[val_idx] = torch.tensor(project_l2_ball(Z[val_idx].data.cpu().numpy()), requires_grad=True).to(device)       
        
        losses.append(loss.item())
        progress.set_postfix({'loss': np.mean(losses[-99:])})
        progress.update()
        
        progress.close()
        
        # visualize reconstructions
        G_zi = train_g(val_idx)
    
        if epoch%50 == 0:
            imsave('./low_shots_pk/%s_find_z_ep_%03d.png' % ('pk', epoch), 
                make_grid(G_zi.data.cpu() / 2. + 0.5, nrow=8).numpy().transpose(1, 2, 0))
        
    # save the network parameter 
    torch.save(train_decoder.state_dict(), train_z_path)
    return target_z
```

```python
epoch_for_train_z = 1250
target_z_hat = find_z(1, epoch_for_train_z, val_loader, idx_val.numpy(), train_g, optimizer, target_z, train_decoder, 
                      "./low_shots_pk/checkpoint/current_checkpoint_train_z.pt")
```

```python
# jointly train
for param in train_decoder.parameters():
    param.requires_grad = True
    
loss_fn = nn.MSELoss()
learning_rate = 1e-4

epoch_for_joint_train = 350# 350

optimizer = torch.optim.Adam(joint_train_g.parameters(), lr=learning_rate)
```

```python
# useing train_decoder and targert_z_hat as warm start
def joint_train(start_epoch, n_epochs, val_loader, val_idx, joint_train_g, optimizer, target_z_hat, train_decoder, joint_train_path):

    train_g.train()
    for epoch in range(start_epoch, n_epochs+1):
        losses = []
        progress = tqdm(total=1, desc='epoch % 3d' % epoch) # for one target image
        
        G_z = joint_train_g(val_idx)
        
        loss = loss_fn(G_z, Xi_val)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # project
        Z[val_idx] = torch.tensor(project_l2_ball(Z[val_idx].data.cpu().numpy()), requires_grad=True).to(device)       
        
        losses.append(loss.item())
        progress.set_postfix({'loss': np.mean(losses[-99:])})
        progress.update()
        
        progress.close()
        
        # visualize reconstructions
        G_zi = joint_train_g(val_idx)
    
        if epoch%10 == 0:
            imsave('./low_shots_pk/%s_final_epoch_%03d.png' % ('pk', epoch), 
                make_grid(G_zi.data.cpu() / 2. + 0.5, nrow=8).numpy().transpose(1, 2, 0))
        
    # save the network parameter 
    torch.save(train_decoder.state_dict(), joint_train_path)
    return target_z
```

```python
target_z_final = joint_train(1, epoch_for_joint_train, val_loader, idx_val.numpy(), joint_train_g, optimizer, target_z_hat, train_decoder, 
                      "./low_shots_pk/checkpoint/current_checkpoint_final.pt")
```
